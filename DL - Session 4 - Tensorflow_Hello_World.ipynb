### Create a tensor

```python
hello = tf.constant('Hello World')
```

```python
print(hello)
```

```python
print(hello.numpy())
```

### Tensor name and shape

```python
hello_2 =  tf.constant('Hello World', shape=[6,2])
print(hello_2)
```

```python
hello_2.numpy()
```

### Adding two tensors

```python
a = tf.constant([1.2])
b = tf.constant([3.4])
c = tf.add(a,b)
print(c)
```

```python
d = 1.2
e = 3.4
f = tf.add(d,e)
print(f)
```
